# Cumulative Density Functions (CDFs)

Cumulative Density Functions (CDFs) are an important concept in probability theory and statistics. Let's break down the various aspects of CDFs:

## Definition:
The Cumulative Density Function (CDF) of a random variable X, denoted as F(x), is a function that gives the probability that X will take a value less than or equal to x. In mathematical terms, for a random variable X,

$$ F(x) = P(X \leq x) $$

For discrete random variables, the CDF is found by summing up the probabilities:

$$ F(x) = \sum_{k \leq x} P(X = k) $$

For continuous random variables, the CDF is the integral of the probability density function (pdf):

$$ F(x) = \int_{-\infty}^{x} f(t) dt $$

where $f(t)$ is the probability density function of $X$.

## Pros:
1. **Complete Information**: CDFs provide complete information about the distribution of a random variable.
2. **Non-decreasing**: CDFs are always non-decreasing, which makes them easier to analyze.
3. **Universality**: CDFs exist for any random variable, whether it is discrete, continuous, or mixed.

## Cons:
1. **Less Intuitive**: For some, CDFs might be less intuitive to understand compared to probability density functions or probability mass functions.
2. **Not Informative for Density**: In the case of continuous random variables, CDFs do not provide information about the density at a specific point.

## Examples:
1. For a standard normal distribution, the CDF at $x=0$ is 0.5, indicating that there is a 50% chance that a random variable from this distribution will be less than or equal to 0.
2. For a uniform distribution on the interval [0, 1], the CDF $F(x)$ is equal to x for $0 <= x <= 1$.

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.stats import norm

# Define a range of x values
x = np.linspace(-3, 3, 1000)

# Define the mean and standard deviation for the normal distribution
mu = 0
sigma = 1

# Calculate the CDF
cdf = norm.cdf(x, mu, sigma)

# Plot the CDF
plt.plot(x, cdf, label='CDF')
plt.xlabel('x')
plt.ylabel('F(x)')
plt.title('Cumulative Density Function')
plt.legend()
plt.grid()
plt.show()
```
